```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()

#to convert String to Dictionory (loads)
record = json.loads(r)
```

```python
print("****Inventory Management System for a Stationary Shop****\n\n")

record
```

```python
#to add new product in the Inventory

prod_id = str(input("Enter product id:"))
name = str(input("Enter name: "))
cn = str(input("Enter Brand name: "))
pr = int(input("Enter price: "))
qn = int(input("Enter quantity: "))

record[prod_id] = {'name': name, 'company_nm' : cn, 'price': pr, 'quantity': qn}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

```python
print("Updated record:--\n\n")

record
```

```python
#to delete a particular record of a item using Product Id

prod_id = str(input("Enter product id that you want to delete:"))

if (prod_id in record.keys()):
  del record[prod_id]
  print("Item deleted")
else:
  print("No item exists")

print("\nYour updated record:- \n")
record
```
